```python
import pandas as pd
grid_predictions = pd.read_csv('grid_prediction.csv')
grid_predictions = grid_predictions.drop(columns=['Unnamed: 0'])
grid_predictions = grid_predictions.set_index('grid_id')
grid_predictions.head()
```

```python
grids = pd.read_csv('/Users/shivambhandari/Downloads/assignment2/data/county_grids.csv')
grids = grids.set_index('grid_id')
grids.head()
```

```python
joined = grids.join(grid_predictions)
joined.head()
```

```python
joined.to_csv('grid_for_qgis.csv')
```

```python
joined.info()
```

```python
joined['timestamp'] = pd.to_datetime(joined['timestamp'])
joined.head()
```

```python
joined.info()
```

```python
joined_6_to_9 = joined.loc[(joined['timestamp'].dt.hour >= 6) & (joined['timestamp'].dt.hour <= 9)]
joined_6_to_9.head()
```

```python
joined_6_to_9 = joined_6_to_9.drop(columns=['timestamp'])
```

```python
joined_6_to_9 = joined_6_to_9.groupby(['grid_id','lat','lon','geom']).mean()
joined_6_to_9.head()
```

```python
joined_6_to_9.to_csv('joined_6_to_9.csv')
```

```python
joined_16_to_19 = joined.loc[(joined['timestamp'].dt.hour >= 16) & (joined['timestamp'].dt.hour <= 19)]
joined_16_to_19 = joined_16_to_19.drop(columns=['timestamp'])
joined_16_to_19 = joined_16_to_19.groupby(['grid_id','lat','lon','geom']).mean()
joined_16_to_19.head()
```

```python
joined_16_to_19.to_csv('joined_16_to_19.csv')
```

```python
joined_weekday = joined.loc[(joined['timestamp'].dt.dayofweek <= 4) & (joined['timestamp'].dt.dayofweek >= 0)]
joined_weekday = joined_weekday.drop(columns=['timestamp'])
joined_weekday = joined_weekday.groupby(['grid_id','lat','lon','geom']).mean()
joined_weekday.head()
```

```python
joined_weekday.to_csv('joined_weekday.csv')
```

```python
joined_weekend = joined.loc[(joined['timestamp'].dt.dayofweek >= 5) & (joined['timestamp'].dt.dayofweek <= 6)]
joined_weekend = joined_weekend.drop(columns=['timestamp'])
joined_weekend = joined_weekend.groupby(['grid_id','lat','lon','geom']).mean()
joined_weekend.head()
```

```python
joined_weekend.to_csv('joined_weekend.csv')
```
